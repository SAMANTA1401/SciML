```python
import jax
import jax.numpy as jnp
```

```python
x = jnp.array([1.0, 2.0, 3.0])
y = jnp.sin(x)
print(y)
```

Auto-differentiation with

```python
from jax import grad

def f(x):
    return x**2 + 3 * x

df = grad(f)
print(df(2.0))  # Output: 7.0 (2x + 3 at x=2)
```

Just-In-Time Compilation

```python
from jax import jit

@jit
def g(x):
    return jnp.sin(x) + jnp.cos(x)

print(g(1.0))  # Much faster after first call
```

```python
from jax import vmap

def square(x):
    return x * x

x = jnp.array([1., 2., 3., 4.])
print(vmap(square)(x))  # Output: [1., 4., 9., 16.]
```
